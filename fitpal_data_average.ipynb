```python
import pandas as pd
```

```python
df = pd.read_csv('total_and_goal_nutrition.csv')
```

```python
df
```

```python
df['average_sugar_consumed'] = (df['total_sugar'] / df['goal_sugar']) * 100
df['average_sodium_consumed'] = (df['total_sodium'] / df['goal_sodium']) * 100
df['average_carbs_consumed'] = (df['total_carbs'] / df['goal_carbs']) * 100
df['average_protein_consumed'] = (df['total_protein'] / df['goal_protein']) * 100
df['average_fat_consumed'] = (df['total_fat'] / df['goal_fat']) * 100
df['average_calories_consumed'] = (df['total_calories'] / df['goal_calories']) * 100
```

```python
df
```

```python
df2 = df.fillna(0.000000)
```

```python
df2
```
