```python
import pandas as pd
from statistics import mode
```

```python
df = pd.read_parquet('sleep.parquet')
```

```python
empty_rows_count = df.isnull().sum()
```

```python
empty_rows_count
```

```python
most_frequent_wakeup = df['Wake up'].mode().iloc[0]
most_frequent_value = df['Sleep Notes'].mode().iloc[0]
mean_heart_rate = df['Heart rate'].mean()
```

```python
df['Wake up'].fillna(value=most_frequent_wakeup, inplace=True)
df['Sleep Notes'].fillna(value=most_frequent_value, inplace=True)
df['Heart rate'].fillna(value=mean_heart_rate, inplace=True)
```

```python
empty_rows_count = df.isnull().sum()
empty_rows_count
```

```python
mask = df['Sleep Notes'].str.contains('Stressful day', case=False)
mask_worked_out = df['Sleep Notes'].str.contains('Worked out', case=False)
mask_drank_coffee = df['Sleep Notes'].str.contains('Drank coffee', case=False)
mask_drank_tea = df['Sleep Notes'].str.contains('Drank tea', case=False)
mask_ate_late = df['Sleep Notes'].str.contains('Ate late', case=False)
```

```python
df['Stressful day'] = False 
df.loc[mask, 'Stressful day'] = True
df['Worked out'] = False
df.loc[mask_worked_out, 'Worked out'] = True

df['Drank coffee'] = False
df.loc[mask_drank_coffee, 'Drank coffee'] = True

df['Drank tea'] = False
df.loc[mask_drank_tea, 'Drank tea'] = True

df['Ate late'] = False
df.loc[mask_ate_late, 'Ate late'] = True
```

```python
df['Start'] = pd.to_datetime(df['Start'])
df['End'] = pd.to_datetime(df['End'])
```

```python
df['Duration'] = df['End'] - df['Start']
```

```python
df['Duration'] = df['Duration'].dt.components['hours'].astype(str).str.zfill(2) + ':' + \
                df['Duration'].dt.components['minutes'].astype(str).str.zfill(2) + ':' + \
                df['Duration'].dt.components['seconds'].astype(str).str.zfill(2)
```

```python
df.to_parquet('sleep_new.parquet', index=False)
```
