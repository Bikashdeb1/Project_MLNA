```python
import pandas as pd
```

```python
df = pd.read_parquet('final_dataset.parquet')
```

```python
df
```

```python
df_pivot = df.pivot_table(index=['User_id', 'Date', 'meal', 'dishes__name', 'sequence'],
                          columns='dishes__nutritions__name',
                          values='dishes__nutritions__value',
                          aggfunc='first').reset_index()

df_pivot.columns.name = None
```

```python
df_pivot
```

```python
df1 = pd.DataFrame(df_pivot)
```

```python
df1
```

```python
df_final = df1.fillna(0)
```

```python
df_final
```

```python
df_final['dishes__quantity'] = df_final['dishes__name'].str.split(',').str[-1].str.strip()
```

```python
df_final['dishes__name'] = df_final['dishes__name'].str.split(',').str[0].str.strip()
```

```python
df_final.head(10)
```

```python
new_column_order = ['User_id', 'Date', 'meal', 'dishes__name', 'dishes__quantity',
                    'Calories', 'Carbs', 'Protein', 'Sodium', 'Sugar','Fat', 'Mon Fat',
                    'Ply Fat', 'Sat Fat', 'Trn Fat', 'Potass.', 'Calcium', 'Chol', 'Fiber', 'Iron',
                    'Vit A', 'Vit C', 'sequence',]
```

```python
df_final_dataset = df_final[new_column_order].rename(columns={'dishes__name': 'Dishe name',
                                                   'dishes__quantity': 'Dishe Quantity', 'meal':'Meal', 'Carbs':'Carbohydrates',
                                                             'Mon Fat':'Monounsaturated Fat', 'Ply Fat':'Polyunsaturated Fat',
                                                             'Sat Fat':'Saturated Fat', 'Trn Fat':'Trans Fat', 'Potass':'Potassium',
                                                             'Chol':'Cholesterol', 'Vit A':'Vitamin A', 'Vit C':'Vitamin C',
                                                             'sequence':'Sequence'})
```

```python
df_final_dataset.to_csv('final_data_set.csv', index = False)
```
