```python
import pandas as pd
```

```python
df = pd.read_parquet('food_name_nutration_value.parquet')
```

```python
df
```

```python
food_id_mapping = {}
counter = 1

# Iterate over the rows in the DataFrame
for index, row in df.iterrows():
    # Convert the Food_ID value to string
    food_id_str = str(row["Food_ID"])

    # Update the Food_ID value for the row
    if food_id_str.startswith("FNT"):
        new_food_id = f"FNT{str(counter).zfill(3)}"
        food_id_mapping[food_id_str] = new_food_id
        df.at[index, "Food_ID"] = new_food_id
        counter += 1

# Sort the DataFrame based on the index
df = df.sort_index()

# Print the mapping dictionary (optional)
print(food_id_mapping)
```

```python
df
```

```python
df.to_parquet("updated_food_name_nutration_value.parquet", index=False)
```
